```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("airbnb.csv")
```

```python
df.head()
```

```python
df.columns
```

### missing values

```python
print(df.isnull().sum())
```

```python
df.info()
```

### hadling missing values

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'review per month' : 0, 'last review': df['last review']. min()},inplace = True)
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
# drop columns 
df = df.drop(columns = ['license', 'house_rules'], errors= 'ignore')
```

```python
df.head()
```

### remove dollar sign 

```python
df['price'] = df ['price'].replace('[\$,]', '', regex = True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex = True).astype(float)
```

```python
df.head()
```

### remove duplicates

```python
df.drop_duplicates(inplace= True)
```

```python
df.info()
```

### Descriptive Statistics

```python
df.describe()
```

## Visualization

### Distribution of listing price

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'], bins = 50, kde = True, color = 'red')
plt.title("Distribution of listing price")
plt.xlabel('price $')
plt.ylabel('frequency')
plt.show()
```

## Different room types Distributed

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot( x = "room type", data = df, color = "hotpink")
plt.title("room types distributed")
plt.xlabel('room type')
plt.ylabel('count')
plt.show()
```

### lisitng distributed acoss different neighborhood

```python
plt.figure(figsize = (12,8))
sns.countplot(y = 'neighbourhood group', data= df, color="darkgreen", order=df ['neighbourhood group']. value_counts().index)
plt.title('number of listing by neighbourhood group')
plt.xlabel('count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

### relation ship between price and room

```python
plt.figure(figsize = (10,6))
sns.boxplot(x='room type', y ='price', hue='room type', data=df, palette='Set1')
plt.title("Price vs Room Type")
plt.xlabel('room type')
plt.ylabel('price ($)')
plt.legend(title= 'room type')
plt.show()
```

### Number of reviews change over with time 

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind= 'line', color='blue')
plt.title("Number of reviews over time")
plt.xlabel('date')
plt.ylabel('Number of Reviews')
plt.show()
```
